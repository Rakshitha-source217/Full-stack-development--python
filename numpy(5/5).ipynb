```python
import numpy as np
list1 = [10, 20, 30]
list2 = [40, 50, 60]
arr = np.array([list1, list2])
col_sum = np.sum(arr, axis=0)
row_mean = np.mean(arr, axis=1)

print("Column Sums:", col_sum)
print("Row Means:", row_mean)
```
